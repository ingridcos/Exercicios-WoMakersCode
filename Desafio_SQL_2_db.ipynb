```python
import sqlite3, pandas as pd

conn = sqlite3.connect('/content/DesafioSQL2.db.db')
```

```python
tabelas = pd.read_sql(
    "SELECT name FROM sqlite_master WHERE type='table' ORDER BY name;",
    conn
)
tabelas
```

```python
dfs = {}
for nome in tabelas['name']:
    dfs[nome] = pd.read_sql(f"SELECT * FROM {nome};", conn)

# exemplo: ver uma tabela específica
dfs['customers'].head()
dfs['order_items'].head()
dfs['orders'].head()
dfs['products'].head()
```
