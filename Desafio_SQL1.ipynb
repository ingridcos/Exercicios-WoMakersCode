```python
import sqlite3
import pandas as pd
```

```python
alunos = pd.read_csv("/content/ALUNOS.csv")
cursos = pd.read_csv("/content/CURSOS.csv")
matriculas = pd.read_csv("/content/MATRICULAS.csv")
```

```python
print("ALUNOS:")
display(alunos)
print("CURSOS:")
display(cursos)
print("MATRICULAS:")
display(matriculas)
```

```python
conn = sqlite3.connect("/content/Desafio-SQL1.db")
cursor = conn.cursor()
```

```python
# Criar tabelas
cursor.execute("""
CREATE TABLE IF NOT EXISTS ALUNOS (
    ID_ALUNO INT PRIMARY KEY,
    NOME_ALUNO VARCHAR(150),
    EMAIL VARCHAR(100),
    DATA_INGRESSO DATE
);
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS CURSOS (
    ID_CURSO INT PRIMARY KEY,
    NOME_CURSO VARCHAR(100),
    DEPARTAMENTO VARCHAR(100),
    CREDITOS INT
);
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS MATRICULAS (
    ID_MATRICULA INT PRIMARY KEY,
    ID_ALUNO INT,
    ID_CURSO INT,
    NOTA_FINAL DECIMAL(4,2),
    FOREIGN KEY (ID_ALUNO) REFERENCES ALUNOS(ID_ALUNO),
    FOREIGN KEY (ID_CURSO) REFERENCES CURSOS(ID_CURSO)
);
""")
```

```python
#Inserir os dados dos CSVs nas tabelas
alunos.to_sql('ALUNOS', conn, if_exists='replace', index=False)
cursos.to_sql('CURSOS', conn, if_exists='replace', index=False)
matriculas.to_sql('MATRICULAS', conn, if_exists='replace', index=False)

conn.commit()

print("Tabelas criadas e dados importados com sucesso!")
```

```python
query = """
SELECT * FROM ALUNOS;
"""
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT
    A.NOME_ALUNO,
    ROUND(AVG(M.NOTA_FINAL), 2) AS MEDIA
FROM ALUNOS A
JOIN MATRICULAS M ON A.ID_ALUNO = M.ID_ALUNO
GROUP BY A.NOME_ALUNO
ORDER BY MEDIA DESC
LIMIT 1;
"""
pd.read_sql_query(query, conn)
```

```python
# Mostrar as tabelas
print("Tabelas no banco:")
tables = cursor.execute("SELECT name FROM sqlite_master WHERE type='table';").fetchall()
for t in tables:
    print("-", t[0])
```
