**1)** Ao visualizar a base você percebeu que seria melhor alterar o nome de
algumas colunas. Mude o ‘ID’ para ‘Identificador’, corrija o nome da coluna
que indica a pressão sanguínea, mude a coluna ‘Ocupação’ para ‘Profissão’,
a coluna ‘Categoria BMI’ está em parte em inglês, substitua para ‘Categoria
IMC’.


```python
df = df.rename(columns={
    'ID': 'Identificador',
    'Ocupação': 'Profissão',
    'Pressão sanguíneaaaa': 'Pressão sanguínea',  # Corrigindo nome
    'Categoria BMI': 'Categoria IMC'
})

print("Colunas renomeadas:")
print(df.columns.tolist())
print("\n")
```

**2.** Qual é a média, a moda e a mediana de horas de sono para cada uma
das profissões? [‘mean’, np.median, pd.Series.mode]


```python
sono_por_profissao = df.groupby('Profissão')['Duração do sono'].agg(
    Média='mean',
    Moda=lambda x: x.mode()[0] if len(x.mode()) > 0 else np.nan,
    Mediana='median'
).round(2)   # arredonda para 2 casas

print("Estatísticas de sono por profissão:")
print(sono_por_profissao, "\n")
```

**4.** De acordo com os dados, advogar ou ser representante de vendas faz

**8.** Quem tem maior pressão sanguínea média, homens ou mulheres?

**9.** É predominante entre os participantes dormir 8 horas por dia
(considere usar Moda como medida)?


```python
moda_sono = df['Duração do sono'].mode()[0]
print(f"Moda das horas de sono: {moda_sono:.1f} horas")
print("   Predominante: " + ("Sim, 8 horas" if abs(moda_sono - 8) < 0.5 else "Não"))
print("\n")
```

**10.** Pessoas com frequências cardíacas acima de 70 dão mais passos que
pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python
df['FC_maior_70'] = df['Frequência cardíaca'] > 70
```

```python
passos_fc = df.groupby('FC_maior_70')['Passos diários'].mean().round(0)
```

```python
print("Média de passos por frequência cardíaca:")
print("   FC <= 70:", f"{passos_fc[False]:,.0f}")
print("   FC > 70: ", f"{passos_fc[True]:,.0f}")
```

```python
if passos_fc[True] > passos_fc[False]:
    print("   FC > 70 dá mais passos? Sim")
else:
    print("   FC > 70 dá mais passos? Não")
```
